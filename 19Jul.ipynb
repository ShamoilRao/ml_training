## MultiIndex
Multi index allows you to select more than one column and row in any index

```python
outside = ['R1','R1','R1','R2','R2', 'R2']
inside = [1,2,3,1,2,3]

# zip takes two or more iterable/container and combine them to reaturn single iterable/container
hier_index = list(zip(outside, inside))

print(f"hier_index: {hier_index}\n")

hier_index = pd.MultiIndex.from_tuples(hier_index)

print(f"hier_index after multiindex: {hier_index}")

df=pd.DataFrame(np.random.randn(6,2), hier_index, ['A', 'B']) 
print(df)

print(df.loc['R1'])
```

```python
fig = matplotlib.pyplot.figure(dpi=200)

a = 5
b = -4
c = -10

x = numpy.linspace(-10, 10, num=1000)
y = a*x**3 + b*x + c + b*x**2

ax = fig.add_axes(rect=(0.5,0.5,0.5,0.5))
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.plot(x,y)
ax.grid(True)
matplotlib.pyplot.show()
```

### figure.plt.subplot()
- Subplots allow numerous plots to be displayed on the same matplotlib figure.plt.subplot()

```python
fig = matplotlib.pyplot.figure(figsize=(10, 6))

# subplot1
fig1 = fig.add_subplot(2,2,1, fc='r')
fig1.plot([1,2,5,7])

# subplot2
fig2 = fig.add_subplot(2,2,2, fc='y')
fig2.plot([1,2,3,5,0])

# subplot2
fig3 = fig.add_subplot(2,2,3)
fig3.plot([1,2,3,5,0])

fig4 = fig.add_subplot(2,2,4, fc='b')
fig.set_edgecolor('r')
fig4.plot([1,2,3,5,0])
fig4.set
```

### Generate plot for matrix 

- By default matrix use coloum major to store in memeory

```python
data = numpy.matrix([[10,20,30,40], [100,200,300,400]])
print(f"matrix: \n{data}")

matplotlib.pyplot.plot(data)
matplotlib.pyplot.show()

data_transpose = data.transpose()
print(f"tranpose: \n{data_transpose}")
matplotlib.pyplot.plot(data_transpose)
matplotlib.pyplot.show()
```

## What is the difference between historgarm and bar chart

- bar chart is used for show frequecy of categorical data, it has categories
- historgram is used for frequency distribution of continuous data, it has ranges of intervals

### Histogram

- used to represent numerical data
- __bin__ = no of bins is used in historgram
- __range__ = this paramerts specifies the range


```python
data = np.random.randint(10,50, size=10)
print(f"data: {data}")
matplotlib.pyplot.hist(data, bins=10, color='red')

matplotlib.pyplot.xlabel("values")
matplotlib.pyplot.ylabel("frequency")

```

# Matplotlib
- Popular plotting library
- It give you contral over every saspect of figure
- It was designed to give similar feel to matlab graphical plotting
- It works well with pandas and numpy

```python
import pandas
import numpy
import matplotlib.pyplot
```

```python
numpy.random.randint(50, size=10)
```

```python
x = np.linspace(0,5,11)
print(f"x: {x}")

y = x ** 2

a = numpy.random.randint(50, size=10)
print(f"a: {a}")

b = numpy.random.randint(20,400, size=15)
print(f"b: {b}")

matplotlib.pyplot.plot(a)
matplotlib.pyplot.plot(b)
matplotlib.pyplot.show() # differnce between show and plot?
```

```python
data = [10, 20, 30, 40]

matplotlib.pyplot.plot(data)
matplotlib.pyplot.show()
    
for i in range(10):
    data.append(i ** 3)
    matplotlib.pyplot.plot([i,i+1])
    matplotlib.pyplot.show() # comment it you will see something
```

### There are two ways to create Maplotlib plot
1. Functional method
2. Object-oriented method

```python
matplotlib.pyplot.plot(x,y)
matplotlib.pyplot.xlabel("X Axis")
matplotlib.pyplot.ylabel("Y Axis")
matplotlib.pyplot.title("Title")

print(x,y)
```

```python
a = numpy.random.randint(5000, size= 12)

matplotlib.pyplot.plot(a, c=(.9, .0, .0), marker='o' , label="Label")
# matplotlib.pyplot.savefig(fname='line.pdf') # also can try line.pdf
```

### If you want to store figure after displying you need hold on the figure instance

```python
a = numpy.random.randint(50, size=10)
b = numpy.random.randint(50, size=10)

matplotlib.pyplot.plot(a,b,':',c='r')
matplotlib.pyplot.show()
matplotlib.pyplot.plot(a,b,'*',c='r')
matplotlib.pyplot.show()
matplotlib.pyplot.plot(a,b,'-*',c='r')
matplotlib.pyplot.show()
matplotlib.pyplot.plot(a,b,'--',c='r')
matplotlib.pyplot.show()

matplotlib.pyplot.plot(a,b,'-.',c='r')
matplotlib.pyplot.show()
```

```python
sales2022 = numpy.random.randint(500000, size=12)

matplotlib.pyplot.plot(sales2022, c='r', label='2022')

matplotlib.pyplot.xlabel("Months")
matplotlib.pyplot.ylabel("Sales")
matplotlib.pyplot.grid(True)
matplotlib.pyplot.title("Sales Prediction of 2022 and 2023")
matplotlib.pyplot.xticks(range(1,6), range(200,205))


sales2023 = numpy.random.randint(500000, size=12)
print(f"sales 2023: {sales2023}")

matplotlib.pyplot.plot(sales2023, color='g', label=2023)

matplotlib.pyplot.legend() # display data label
# matplotlib.pyplot.legend(loc='best') # by defalult value of loc
matplotlib.pyplot.legend(loc='lower right') # label position
```

### Marker are used to display data points
possible markers:
- 'o' 	Circle 	
- '*' 	Star 	
- '.' 	Point 	
- ',' 	Pixel 	
- 'x' 	X 	
- 'X' 	X (filled) 	
- '+' 	Plus 	
- 'P' 	Plus (filled) 	
- 's' 	Square 	
- 'D' 	Diamond 	
- 'd' 	Diamond (thin) 	
- 'p' 	Pentagon 	
- 'H' 	Hexagon

```python
data = numpy.random.randint(40000, size=10)

matplotlib.pyplot.plot(data, marker='p')
```
